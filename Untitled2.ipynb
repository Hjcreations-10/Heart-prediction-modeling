```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,mean_squared_error,mean_absolute_error,r2_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.svm import SVR
from datetime import datetime
from sklearn.cluster import AgglomerativeClustering
```

to calculate statistical values

```python
df['ldl']=df['ldl'].fillna(df['ldl'].mean())
print(f'mean:{df}')
df['hdl']=df['hdl'].fillna(df['hdl'].median())
print(f'meadian:{df}')
df['systolic_bp']=df['systolic_bp'].fillna(df['systolic_bp'].mode())
print(f'mode:{df}')
```

testing and training

```python
x=df[['ldl','hdl','systolic_bp']]
y=df['heart_attack']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape,x_test.shape)
```

feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

SVM model

```python
model=SVR()
model.fit(x_train_scaled,y_train)
predicts=model.predict(x_test_scaled)
mse=mean_squared_error(y_test,predicts)
mbe=mean_absolute_error(y_test,predicts)
r2=r2_score(y_test,predicts)
print(f'mean squared error:{mse},mean absolute error:{mbe},r square={r2}')
```

DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
model=dt.fit(x_train_scaled,y_train)
predicts=model.predict(x_test_scaled)
cf=confusion_matrix(y_test,predicts)
cr=classification_report(y_test,predicts)
ac=accuracy_score(y_test,predicts)
print(f'confusion matrix:{cf},classification report:{cr},accuracy_score={ac}')
```

LogisticRegression

```python
lr= LogisticRegression()
model=lr.fit(x_train_scaled,y_train)
predicts=model.predict(x_test_scaled)
cr=classification_report(y_test,predicts)
ac=accuracy_score(y_test,predicts)
print(f'classification report:{cr},accuracy_score={ac}')
```

Sample vs Prediction

```python
plt.figure(figsize=(12,6))
plt.scatter(range(len(y_test)),y_test,color='blue',label='Actual')
plt.scatter(range(len(predicts)),predicts,color='red',label='predicted')
plt.title('Heart Attack Prediction')
plt.xlabel('Sample')
plt.ylabel('Heart Attack')
plt.legend()
plt.grid()
plt.show()
```

VotingRegressor

```python
from sklearn.ensemble import VotingRegressor
estimators=[('LR',LinearRegression()),('rf',RandomForestRegressor()),('svm',SVR())]
vr= VotingRegressor(estimators=estimators)
vr.fit(x_train_scaled,y_train)
predicts=vr.predict(x_test_scaled)
mse=mean_squared_error(y_test,predicts)
mbe=mean_absolute_error(y_test,predicts)
r2=r2_score(y_test,predicts)
print(f'mean squared error:{mse},mean absolute error:{mbe},r square={r2}')
```

STACKINGCLASSIFIER

```python
from sklearn.ensemble import StackingClassifier
estimators=[('lr',LogisticRegression()),('rf',RandomForestClassifier())]
SC= StackingClassifier(estimators=estimators)
model=SC.fit(x_train_scaled,y_train)
predicts=model.predict(x_test_scaled)
cf=confusion_matrix(y_test,predicts)
cr=classification_report(y_test,predicts)
ac=accuracy_score(y_test,predicts)
print(f'confusion matrix:{cf},classification report:{cr},accuracy_score={ac}')
```
